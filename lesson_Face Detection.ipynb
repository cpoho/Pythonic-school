# Раcпознавание лиц



Одна из самых обсуждаемых тем в последнее время - это распознавание лиц. 
Государства активно внедряют системы распознавания лиц для поиска приступников, контроля за гражданами и т.д.
Активисты борятся за право перемещаться по городу без излишнего внимания к собственным персонам. Причем достигают успеха - в США уже в трех штатах запрещено использование силовыми структурами систем распознавания лиц; Евросоюз рассматривает возможность запрета таких систем в публичных местах на пять лет.

С другой стороны алгоритмы распознавания лиц активно используются в таких технологиях как AR, нейросети. Без них невозможно было бы сделать ни масочки в инстаграм, ни задипфейчить звезд и политиков.


Существует множество подходов к распознаванию лиц, мы рассмотрим два алгоритма - алгоритм Виола-Джонса и гистограммы направленных градиентов.


## Виола-Джонс

Несмотря на то, что разработки по распознованию лиц велись еще с 1970-х годов, настоящий прорыв произошел в 2001 году, когда Полом Виолой и Майклом Джонсом был предложен метода каскадного бустинга. Революционность этого метода заключалась в его высокой надежности и производительности - даже на обычном компьютере тех времен можно было обрабатывать более десяти кадров в секунду. Сейчас это один из самых популярных алгоритмов, используемый в телефонах, фотоаппаратах, камерах наблюдения.

Алгоритм можно разбить на четыре шага
1. Изображение переводится в чернобелый формат, таким образом превращая его в матрицу яркости (Y = 0.299*R+0.587*G+0.114*B)



2. Применяем к изображению одну из прямоугольных масок (также известных как признаки Хаара). Проходимся этой маской по всему изображению, меняя ее положение и размер



3. Вычисляем признаки. Для этого из суммы значений, попавших в белую часть маски, вычитаем сумму значений из черной части. Если разность белых и черных областей оказалась больше заданного порога, используем эту область для дальнейшей работы, если нет - считаем, что здесь лица нет.



4. Снова возвращаемся к шагу два. Но в этот раз берем следующую маску и проходимся ею по отобранным в этой итерации областям. Такая каскадность позволяет снизить общую ошибку. Такой подход называется бустингом - в алгоритме Виола-Джонса используется метод машинного обучения AdaBoost, который создает сильный классификатор из ряда слабых классификаторов (признаков Хаара).

В конечном итоге алгоритм выдает область, в которой найдено лицо.
